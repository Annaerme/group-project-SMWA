## 1. VADER

VADER (Valence Aware Dictionary and sEntiment Reasoner) is a rule-based model that handles capitalisation, punctuation, and social media conventions. Returns a compound score ∈ [−1, +1].

```python
vader = SentimentIntensityAnalyzer()

def vader_scores(text):
    scores = vader.polarity_scores(str(text))
    return scores['compound'], scores['pos'], scores['neg'], scores['neu']

(
    df['vader_compound'],
    df['vader_pos'],
    df['vader_neg'],
    df['vader_neu'],
) = zip(*df['title_clean'].map(vader_scores))

# Classify as positive / negative / neutral (standard VADER thresholds)
def vader_label(compound):
    if compound >=  0.05: return 'positive'
    if compound <= -0.05: return 'negative'
    return 'neutral'

df['vader_label'] = df['vader_compound'].map(vader_label)

print('VADER score distribution (overall):')
print(df['vader_label'].value_counts(normalize=True).mul(100).round(1).to_string())
print(f'\nMean compound score : {df["vader_compound"].mean():+.4f}')
print(f'Std  compound score : {df["vader_compound"].std():.4f}')
```

```python
# ── Distribution of compound scores by leaning ──────────────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharey=True)

for ax, (leaning, color) in zip(axes, LEANING_COLORS.items()):
    sub = df[df['leaning'] == leaning]['vader_compound']
    ax.hist(sub, bins=40, color=color, edgecolor='white', alpha=0.85)
    ax.axvline(sub.mean(), color='black', linestyle='--', linewidth=1.5,
               label=f'mean = {sub.mean():+.3f}')
    ax.axvline(0, color='grey', linestyle=':', linewidth=1)
    ax.set_title(leaning, fontsize=12, fontweight='bold')
    ax.set_xlabel('VADER compound score')
    ax.legend(fontsize=9)
    ax.grid(axis='y', alpha=0.3)

axes[0].set_ylabel('Number of articles')
fig.suptitle('VADER Compound Score Distribution by Political Leaning',
             fontsize=14, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()
```

## 2. TextBlob

**What it is:** TextBlob uses the Pattern library lexicon to score text on polarity ∈ [−1, +1] and subjectivity ∈ [0, 1].

```python
from textblob import TextBlob

def tb_scores(text):
    blob = TextBlob(str(text))
    return blob.sentiment.polarity, blob.sentiment.subjectivity

tb_results = df['title_clean'].fillna('').apply(tb_scores)
df['tb_polarity']     = [r[0] for r in tb_results]
df['tb_subjectivity'] = [r[1] for r in tb_results]

def tb_label(p):
    if p >  0.05: return 'positive'
    if p < -0.05: return 'negative'
    return 'neutral'

df['tb_label'] = df['tb_polarity'].map(tb_label)

print('TextBlob score distribution (overall):')
print(df['tb_label'].value_counts(normalize=True).mul(100).round(1).to_string())
print(f'\nMean polarity     : {df["tb_polarity"].mean():+.4f}')
print(f'Mean subjectivity : {df["tb_subjectivity"].mean():.4f}')
```

## 3. NRCLex — Emotion Detection

NRCLex maps words to 8 discrete emotions: *fear, anger, anticipation, trust, surprise, sadness, disgust, joy*, plus overall positive/negative labels.

```python
EMOTIONS = ['fear', 'anger', 'anticipation', 'trust',
            'surprise', 'sadness', 'disgust', 'joy']

def nrc_scores(text):
    """Return NRCLex affect frequencies for a headline (8 emotions + pos/neg)."""
    freq = NRCLex(str(text)).affect_frequencies
    return {e: freq.get(e, 0.0) for e in EMOTIONS + ['positive', 'negative']}

print('Running NRCLex on all headlines ...')
nrc_df = df['title_clean'].apply(nrc_scores).apply(pd.Series)

for col in nrc_df.columns:
    df[f'nrc_{col}'] = nrc_df[col].values

df['nrc_sentiment'] = df['nrc_positive'] - df['nrc_negative']
nrc_cols = [f'nrc_{e}' for e in EMOTIONS]
print('Done.  Sample output:')
df[['title'] + nrc_cols].head(4)
```

```python
# ── Mean emotion profile per leaning ──────────────────────────────────────────
# EMOTION_COLORS comes from c-imports (house_style constants)

fig, axes = plt.subplots(1, 3, figsize=(16, 5), sharey=True)

for ax, (leaning, lean_color) in zip(axes, LEANING_COLORS.items()):
    sub = df[df['leaning'] == leaning]
    means = sub[[f'nrc_{e}' for e in EMOTIONS]].mean()
    means.index = EMOTIONS

    colors = [EMOTION_COLORS[e] for e in EMOTIONS]
    bars = ax.bar(EMOTIONS, means.values, color=colors, edgecolor='white', linewidth=0.5)

    for bar, val in zip(bars, means.values):
        ax.text(bar.get_x() + bar.get_width() / 2, val + 0.001,
                f'{val:.3f}', ha='center', va='bottom', fontsize=7.5)

    ax.set_title(leaning, fontsize=12, fontweight='bold', color=lean_color)
    ax.set_xticklabels(EMOTIONS, rotation=40, ha='right', fontsize=9)
    ax.set_ylabel('Mean affect frequency')
    ax.grid(axis='y', alpha=0.3)

fig.suptitle('Mean NRCLex Emotion Profile by Political Leaning', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 5. Feature Engineering & Export

Aggregate VADER and NRCLex scores into a daily basetable by political leaning and save for downstream modelling.

```python
# ── 5a. Pivot helper ───────────────────────────────────────────────────────────────────
def pivot_leaning(long_df, value_cols):
    """Pivot daily long-format DataFrame to wide format with leaning suffixes."""
    frames = []
    for col in value_cols:
        for leaning, short in LEANING_SHORT.items():
            sub = long_df[long_df['leaning'] == leaning][['date', col]].copy()
            sub = sub.rename(columns={col: f'{col}_{short}'})
            frames.append(sub.set_index('date'))
    return pd.concat(frames, axis=1)
```

```python
# ── VADER daily aggregations ──────────────────────────────────────────────────
df['vader_is_pos'] = (df['vader_label'] == 'positive').astype(float)
df['vader_is_neg'] = (df['vader_label'] == 'negative').astype(float)

vader_agg = (
    df.groupby(['date', 'leaning'])
    .agg(
        vader_compound_mean=('vader_compound', 'mean'),
        vader_compound_std =('vader_compound', 'std'),
        vader_pos_share    =('vader_is_pos',   'mean'),
        vader_neg_share    =('vader_is_neg',   'mean'),
    )
    .reset_index()
)

vader_wide = pivot_leaning(
    vader_agg,
    ['vader_compound_mean', 'vader_compound_std', 'vader_pos_share', 'vader_neg_share']
)

print(f'VADER feature matrix : {vader_wide.shape}')
vader_wide.head(3)
```

```python
# ── 5c. NRCLex daily aggregations ───────────────────────────────────────────────────
nrc_agg_spec = {f'nrc_{e}': (f'nrc_{e}', 'mean') for e in EMOTIONS}

nrc_agg = (
    df.groupby(['date', 'leaning'])
    .agg(**nrc_agg_spec)
    .reset_index()
)

nrc_wide = pivot_leaning(nrc_agg, [f'nrc_{e}' for e in EMOTIONS])

print(f'NRCLex feature matrix : {nrc_wide.shape}')
nrc_wide.head(3)
```

```python
# ── Combine & save ───────────────────────────────────────────────────────────
sentiment_features = (
    vader_wide
    .join(nrc_wide, how='outer')
    .sort_index()
    .reset_index()
)

out_path = SILVER / 'sentiment_features_newspapers.csv'
sentiment_features.to_csv(out_path, index=False)

n_feat = sentiment_features.shape[1] - 1
print(f'Saved  -> {out_path}')
print(f'Shape  : {sentiment_features.shape}  ({n_feat} features + date)')
print(f'Dates  : {sentiment_features["date"].min().date()} -> {sentiment_features["date"].max().date()}')
print(f'\nFeature groups:')
print(f'  VADER  (compound mean, std, pos/neg share) x 3 leanings : {4 * 3:>3} columns')
print(f'  NRCLex (8 emotions x mean) x 3 leanings                 : {8 * 3:>3} columns')
print(f'  Total (excl. date)                                       : {n_feat:>3} columns')

sentiment_features.head(3)
```

```python
# 6. Repair missing sentiment dates (incl. October check) and overwrite export

import pandas as pd
import numpy as np
from pathlib import Path

def _first_existing(paths):
    for p in paths:
        p = Path(p)
        if p.exists():
            return p
    return Path(paths[0])

# Resolve paths robustly (works whether CWD is notebook folder or project root)
data_candidates = [
    '../../Data/2_Silver/Newspapers/mediacloud_articles_clean.csv',
    'Data/2_Silver/Newspapers/mediacloud_articles_clean.csv',
    '../Data/2_Silver/Newspapers/mediacloud_articles_clean.csv',
]
silver_candidates = [
    '../../Data/2_Silver/Newspapers',
    'Data/2_Silver/Newspapers',
    '../Data/2_Silver/Newspapers',
]

if 'DATA' not in globals() or not Path(DATA).exists():
    DATA = _first_existing(data_candidates)
else:
    DATA = Path(DATA)

if 'SILVER' not in globals() or not Path(SILVER).exists():
    SILVER = _first_existing(silver_candidates)
else:
    SILVER = Path(SILVER)

if 'LEANING_SHORT' not in globals():
    LEANING_SHORT = {
        'Democratic': 'dem',
        'Republican': 'rep',
        'Center/Unknown': 'cen',
    }

if 'EMOTIONS' not in globals():
    EMOTIONS = ['fear', 'anger', 'anticipation', 'trust', 'surprise', 'sadness', 'disgust', 'joy']

out_path = SILVER / 'sentiment_features_newspapers.csv'

# Expected sentiment feature columns
expected_cols = []
for base in ['vader_compound_mean', 'vader_compound_std', 'vader_pos_share', 'vader_neg_share']:
    for short in LEANING_SHORT.values():
        expected_cols.append(f'{base}_{short}')
for emo in EMOTIONS:
    for short in LEANING_SHORT.values():
        expected_cols.append(f'nrc_{emo}_{short}')

# ------------------------------------------------------------------
# Strategy A: rebuild from df if df + sentiment columns are available
# Strategy B: fallback to existing sentiment_features file only
# ------------------------------------------------------------------
source_mode = None

required_df_cols = ['date', 'leaning', 'vader_compound', 'vader_label'] + [f'nrc_{e}' for e in EMOTIONS]

if 'df' not in globals():
    if Path(DATA).exists():
        try:
            df = pd.read_csv(DATA, parse_dates=['date'])
        except Exception:
            pass

can_rebuild_from_df = ('df' in globals()) and all(c in df.columns for c in required_df_cols)

if can_rebuild_from_df:
    source_mode = 'rebuild_from_df'

    work = df.copy()
    work['date'] = pd.to_datetime(work['date'], errors='coerce').dt.normalize()
    work = work.dropna(subset=['date']).copy()
    work = work[work['leaning'].isin(LEANING_SHORT.keys())].copy()

    def pivot_leaning_local(long_df, value_cols):
        frames = []
        for col in value_cols:
            for leaning, short in LEANING_SHORT.items():
                sub = long_df[long_df['leaning'] == leaning][['date', col]].copy()
                sub = sub.rename(columns={col: f'{col}_{short}'})
                frames.append(sub.set_index('date'))
        return pd.concat(frames, axis=1)

    all_dates = pd.date_range(work['date'].min(), work['date'].max(), freq='D')
    all_leanings = list(LEANING_SHORT.keys())
    base_grid = pd.MultiIndex.from_product([all_dates, all_leanings], names=['date', 'leaning']).to_frame(index=False)

    # VADER
    work['vader_is_pos'] = (work['vader_label'] == 'positive').astype(float)
    work['vader_is_neg'] = (work['vader_label'] == 'negative').astype(float)

    vader_agg = (
        work.groupby(['date', 'leaning'])
        .agg(
            vader_compound_mean=('vader_compound', 'mean'),
            vader_compound_std=('vader_compound', 'std'),
            vader_pos_share=('vader_is_pos', 'mean'),
            vader_neg_share=('vader_is_neg', 'mean'),
        )
        .reset_index()
    )
    vader_agg = base_grid.merge(vader_agg, on=['date', 'leaning'], how='left')
    vader_wide = pivot_leaning_local(
        vader_agg,
        ['vader_compound_mean', 'vader_compound_std', 'vader_pos_share', 'vader_neg_share']
    )

    # NRC
    nrc_cols = [f'nrc_{e}' for e in EMOTIONS]
    nrc_agg = (
        work.groupby(['date', 'leaning'])[nrc_cols]
        .mean()
        .reset_index()
    )
    nrc_agg = base_grid.merge(nrc_agg, on=['date', 'leaning'], how='left')
    nrc_wide = pivot_leaning_local(nrc_agg, nrc_cols)

    sentiment_features = (
        vader_wide
        .join(nrc_wide, how='outer')
        .sort_index()
    )

else:
    source_mode = 'repair_existing_features'
    if not out_path.exists():
        raise FileNotFoundError(
            f'Cannot rebuild from df (missing sentiment columns) and no existing file found at: {out_path}'
        )

    sentiment_features = pd.read_csv(out_path, parse_dates=['date']).set_index('date').sort_index()

# Ensure missing whole dates are explicitly present in the index
if can_rebuild_from_df:
    full_dates = pd.date_range(work['date'].min(), work['date'].max(), freq='D')
else:
    if Path(DATA).exists():
        date_ref = pd.read_csv(DATA, usecols=['date'], parse_dates=['date'])
        date_ref['date'] = pd.to_datetime(date_ref['date'], errors='coerce').dt.normalize()
        date_ref = date_ref.dropna(subset=['date'])
        full_dates = pd.date_range(date_ref['date'].min(), date_ref['date'].max(), freq='D')
    else:
        full_dates = pd.date_range(sentiment_features.index.min(), sentiment_features.index.max(), freq='D')
sentiment_features = sentiment_features.reindex(full_dates)
sentiment_features.index.name = 'date'

# Ensure all expected columns exist
for c in expected_cols:
    if c not in sentiment_features.columns:
        sentiment_features[c] = np.nan

# Identify October missing BEFORE repair
feat_check = sentiment_features.copy().reset_index().rename(columns={'index': 'date'})
feat_check['date'] = pd.to_datetime(feat_check['date'], errors='coerce').dt.normalize()
oct_before = feat_check[(feat_check['date'].dt.year == 2024) & (feat_check['date'].dt.month == 10)].copy()
missing_oct_before = oct_before.loc[oct_before[expected_cols].isna().any(axis=1), 'date'].dt.date.unique().tolist()

print(f'Using DATA: {DATA}')
print(f'Using OUT : {out_path}')
print(f'Source mode: {source_mode}')
print('October days with missing sentiment BEFORE repair:')
print(missing_oct_before if missing_oct_before else 'None')

# Fill missing values smoothly across time
sentiment_features = sentiment_features.sort_index()
for c in expected_cols:
    s = sentiment_features[c].astype(float)
    s = s.interpolate(method='linear', limit_direction='both')
    s = s.ffill().bfill()
    sentiment_features[c] = s

# Final dataframe
sentiment_features = sentiment_features.reset_index().rename(columns={'index': 'date'})
sentiment_features['date'] = pd.to_datetime(sentiment_features['date'], errors='coerce').dt.normalize()

# October missing AFTER repair
oct_after = sentiment_features[(sentiment_features['date'].dt.year == 2024) & (sentiment_features['date'].dt.month == 10)].copy()
missing_oct_after = oct_after.loc[oct_after[expected_cols].isna().any(axis=1), 'date'].dt.date.unique().tolist()

print('\nOctober days with missing sentiment AFTER repair:')
print(missing_oct_after if missing_oct_after else 'None')

# Save
sentiment_features.to_csv(out_path, index=False)
print(f'\nSaved repaired features to: {out_path}')
print(f'Shape: {sentiment_features.shape} (date + {len(expected_cols)} sentiment columns)')

sentiment_features.head(5)
```
