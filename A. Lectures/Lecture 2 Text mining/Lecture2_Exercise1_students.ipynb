````python
# Example: Scraping from Mastodon
from mastodon import Mastodon
import pandas as pd

# Connect to a Mastodon instance (e.g., mastodon.social)
# No authentication needed for public timeline search
m = Mastodon(
    api_base_url='https://mastodon.social'
)

# Search for a hashtag
hashtag = "MachineLearning"  # Replace with your chosen hashtag
limit = 1000

# Get posts
posts = m.timeline_hashtag(hashtag, limit=limit)

# Extract text content
texts = [post['content'] for post in posts]

print(f"Collected {len(texts)} posts")
print(f"\nFirst post example:\n{texts[0][:200]}...")
````

### Clean HTML Tags (for Mastodon)

Mastodon posts often contain HTML tags. Use this helper function to remove them:

````python
import re

def clean_html(text):
    """Remove HTML tags from text."""
    clean = re.sub('<.*?>', '', text)
    return clean

# Apply to your texts if using Mastodon
# texts = [clean_html(t) for t in texts]
````
