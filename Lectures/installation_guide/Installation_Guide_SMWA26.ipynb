## Create a virtual environment

It is recommended to create a virtual environment to install the required packages.
Essentially, a virtual environment is a self-contained directory
that contains a Python installation for a particular version of Python, plus a number of additional packages.
Not using a virtual environment can lead to conflicts between packages and versions.

> **TIP:**  
> Creating virtual environments is a best practice when working with Python projects.  
> It is a useful skill to learn now, since you will use it in the future when you go work inside a company.

We will use conda to create the virtual environment
(you should have conda installed when you installed Anaconda for your jupyter notebook). We will install Python 3.11, since this is compatible with all the packages we will use. Other version like 3.10 and 3.12 are also stable and can also be considered for other environments. 

### VSCode

Follow the steps from this [video](https://www.youtube.com/watch?v=fnb4_MzpZFU) to create a virtual environment.

> **IMPORTANT:**  
> At some point in the video, the instructor uses the command `conda create --name mynewenv`.
> You should use `conda create --name SMWA2026 python=3.11` instead (see below).


### Command Line

To create a virtual environment, run the following command in bash terminal window (not in a notebook window!):

```bash
conda create --name SMWA2026 python=3.11
```

To activate the virtual environment, run the following command:

```bash
conda activate SMWA2026
```

It might be required to restart your notebook and select the correct kernel.
